### Final summarizer with local PDF upload
The following cells expose the summarizer via a local FastAPI endpoint that accepts PDF uploads and returns a consolidated summary.

```python
%pip install pypdf

import io
import os
import threading
from pathlib import Path
from typing import List

import nest_asyncio
from fastapi import FastAPI, File, HTTPException, UploadFile
from fastapi.middleware.cors import CORSMiddleware
from pypdf import PdfReader
import uvicorn

DEFAULT_MODEL = "meta-llama/llama-4-maverick:free"
MAX_CHARS_PER_CHUNK = 4500

def get_client(api_key: str | None = None) -> OpenAI:
    key = api_key or os.getenv("OPENROUTER_API_KEY")
    if not key:
        raise RuntimeError("Set OPENROUTER_API_KEY or pass api_key to get_client().")
    return OpenAI(base_url="https://openrouter.ai/api/v1", api_key=key)

def chunk_text(text: str, max_chars: int = MAX_CHARS_PER_CHUNK) -> List[str]:
    clean_text = "\n".join(line.strip() for line in text.splitlines())
    paragraphs = [p for p in clean_text.split("\n\n") if p]
    if not paragraphs:
        return [clean_text]
    chunks: List[str] = []
    current: List[str] = []
    current_length = 0
    for para in paragraphs:
        para_len = len(para)
        if current and current_length + para_len + 2 > max_chars:
            chunks.append("\n\n".join(current))
            current = [para]
            current_length = para_len
        else:
            current.append(para)
            current_length += para_len + 2
    if current:
        chunks.append("\n\n".join(current))
    return chunks or [clean_text]

def extract_text_from_pdf(pdf_bytes: bytes) -> str:
    reader = PdfReader(io.BytesIO(pdf_bytes))
    pages = [(page.extract_text() or "").strip() for page in reader.pages]
    text = "\n\n".join(filter(None, pages)).strip()
    if not text:
        raise ValueError("No extractable text found in the uploaded PDF.")
    return text

def summarize_text(text: str, model: str = DEFAULT_MODEL, api_key: str | None = None, *, chunks: List[str] | None = None) -> str:
    client = get_client(api_key)
    chunk_list = chunks or chunk_text(text)
    interim: List[str] = []
    for idx, chunk in enumerate(chunk_list, start=1):
        messages = [
            {"role": "system", "content": "You are an expert technical summarizer."},
            {
                "role": "user",
                "content": f"Chunk {idx} of {len(chunk_list)}:\n\n{chunk}"
            }
        ]
        response = client.chat.completions.create(
            model=model,
            messages=messages,
            temperature=0.2,
            max_tokens=700
        )
        interim.append(response.choices[0].message.content.strip())
    if len(interim) == 1:
        return interim[0]
    synthesis_messages = [
        {"role": "system", "content": "You synthesize multiple partial summaries into one cohesive study guide."},
        {
            "role": "user",
            "content": "\n\n".join(interim)
        }
    ]
    final_response = client.chat.completions.create(
        model=model,
        messages=synthesis_messages,
        temperature=0.2,
        max_tokens=700
    )
    return final_response.choices[0].message.content.strip()
```

```python
app = FastAPI(title="Local PDF Summarizer", version="1.0.0")
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

@app.post("/summaries/pdf")
async def summarize_pdf_endpoint(file: UploadFile = File(...), model: str = DEFAULT_MODEL):
    if not file.filename.lower().endswith(".pdf"):
        raise HTTPException(status_code=400, detail="Upload must be a PDF file.")
    pdf_bytes = await file.read()
    try:
        text = extract_text_from_pdf(pdf_bytes)
        chunks = chunk_text(text)
        summary = summarize_text(text, model=model, chunks=chunks)
    except ValueError as exc:
        raise HTTPException(status_code=400, detail=str(exc)) from exc
    except RuntimeError as exc:
        raise HTTPException(status_code=500, detail=str(exc)) from exc
    return {
        "filename": file.filename,
        "model": model,
        "chunk_count": len(chunks),
        "summary": summary
    }
```

```python
_api_thread: threading.Thread | None = None

def launch_local_api(host: str = "127.0.0.1", port: int = 8000) -> threading.Thread:
    global _api_thread
    if _api_thread and _api_thread.is_alive():
        print(f"Local API already running at http://{host}:{port}")
        return _api_thread
    nest_asyncio.apply()
    config = uvicorn.Config(app, host=host, port=port, log_level="info")
    server = uvicorn.Server(config)

    def _run_server() -> None:
        server.run()

    _api_thread = threading.Thread(target=_run_server, daemon=True)
    _api_thread.start()
    print(f"Local API started on http://{host}:{port}")
    print("Interactive docs: http://127.0.0.1:8000/docs")
    return _api_thread

# launch_local_api()
```

```python
import requests

def summarize_pdf_file(pdf_path: Path, model: str = DEFAULT_MODEL, host: str = "127.0.0.1", port: int = 8000) -> dict:
    if not pdf_path.exists():
        raise FileNotFoundError(f"PDF not found at {pdf_path}")
    url = f"http://{host}:{port}/summaries/pdf"
    with pdf_path.open("rb") as stream:
        response = requests.post(
            url,
            data={"model": model},
            files={"file": (pdf_path.name, stream, "application/pdf")},
            timeout=120
        )
    response.raise_for_status()
    return response.json()

# Example usage (adjust the search path or set manual_pdf if needed):
manual_pdf: Path | None = None  # e.g., Path(r"C:\Users\ADMIN\Downloads\paper.pdf")
search_root = Path(r"C:\Users\ADMIN\OneDrive\Desktop\Test")

if manual_pdf and manual_pdf.exists():
    target_pdf = manual_pdf
else:
    pdf_candidates = sorted(search_root.rglob("*.pdf"))
    target_pdf = pdf_candidates[0] if pdf_candidates else None

if target_pdf:
    print(f"Summarizing: {target_pdf}")
    result = summarize_pdf_file(target_pdf)
    print(result["summary"])
else:
    print("No PDFs found. Place a PDF under "f"{search_root.resolve()} or set manual_pdf to the correct file.")
```
